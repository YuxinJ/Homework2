### question 3a

```python
#make community_pop into a dataframe 
community_popcount = pd.DataFrame({'pop': community_pop})
```

```python
#concaatenate the two dataframe
community_crime_pop = pd.concat([community_crimecount,community_popcount], axis=1)
```

```python
community_crime_pop.dtypes
```

```python
def per1000(number):
    return number/1000
community_crime_pop['pop'] = community_crime_pop['pop'].apply(per1000)
```

```python
community_crime_pop.fillna(0,inplace=True)
community_crime_pop[:0]
```

```python
#caculate crime rate 
community_crime_pop['crime_rate']=""
for i in range(1,78):
    rate = community_crime_pop['Crime Count'][i]/community_crime_pop['pop'][i]
    community_crime_pop['crime_rate'][i] = rate
    
```

```python
#result for 3a: display crime rate for each community area
community_crime_pop
```
