```python
def distance((x,y),(z,w)):
    lat = (x-z)*95
    lon = (y-w)*95
    distancesqr = abs(lat)**2 + abs(lon)**2
    distance = distancesqr**0.5
    sentence = 'The distance between these two points is {}km.'.format(distance)
    print(sentence)
```

### question 4d


```python
def distance(row):
    crime_lat = row['Latitude']
    crime_lon = row['Longitude']
    station_lat = row['Station Latitude']
    station_lon = row['Station Longitude']
    lat = (crime_lat - station_lat)*95
    lon = (crime_lon - station_lon)*95
    return (lat**2 + lon**2)**0.5
```

```python
def tofloat(number):
    return float(number)
district['Station Latitude'] = district['Station Latitude'].apply(tofloat)
district['Station Longitude'] = district['Station Longitude'].apply(tofloat)
```

```python
district['distance']=""
district['distance'] = district.apply(distance, axis=1)
```

### question 4e

```python
#drop all data entris if distance is NaN
district = district.dropna(subset=[distance])
```

```python
#turn all entries of distance into integers 
def ins(number):
    return int(number)
district['distance'] = district['distance'].apply(ins)
```

```python
#first aggregate crimes by distance, then calculate crime count for each common distance 
district_distance = district.groupby('distance')
district_distance_count = district_distance['ID'].agg('count')
```

```python
district_distance_count.hist(bins=5,xlabelsize=10)
```
